```python
import pandas as pd
from sklearn.utils import all_estimators
from sklearn.linear_model import LogisticRegressionCV,LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import root_mean_squared_error,classification_report
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import GradientBoostingClassifier,RandomForestClassifier,HistGradientBoostingClassifier
import matplotlib.pyplot as plt
import numpy as np
import koreanize_matplotlib
import os
from collections import Counter
import cv2
```

```python
DF=pd.read_csv('./dog_cat.csv')
testDF=pd.read_csv('./dog_cat_test.csv')
```

```python

DF.drop(columns=['Unnamed: 0'],inplace=True)
testDF.drop(columns=['Unnamed: 0'],inplace=True)
# # a=DF.iloc[4,:].values.reshape(100,100)
# # a=np.uint8(a)
# DF.iloc[4,:].unique()
# # a=np.array(a)
# # print(a)
# # cv2.imshow('g',a)
# # cv2.waitKey()
# # cv2.destroyAllWindows()
```

```python
DF1=DF[DF['speices']=='Dog'].iloc[:744]
DF[DF['speices']=='cat']
DF=pd.concat([DF[DF['speices']=='cat'],DF1])
```

```python
DF1=testDF[testDF['speices']=='Dog'].iloc[:744]
testDF[testDF['speices']=='cat']
testDF=pd.concat([testDF[testDF['speices']=='cat'],DF1])
```

```python
targetDF=DF['speices']
featureDF=DF.iloc[:,:-1]/255
target_testDF=testDF['speices']
feature_testDF=testDF.iloc[:,:-1]/255
```

```python
label=LabelEncoder()
target=label.fit_transform(targetDF)
target_test=label.transform(target_testDF)
# featureDF=featureDF/255
# feature_testDF=feature_testDF/255
featureDF
label.classes_
```

```python
x_train,x_test,y_train,y_test=train_test_split(featureDF,target,stratify=target)
```

```python
name=all_estimators('classifier')
for x,y in name:
    try:
        model=y()
        if 'Logistic' in x or 'SGD' in x or 'MLP' in x:
            model.set_params(max_iter=10000)
        if 'SV' in x:
            model.set_params(max_iter=100000, dual='auto')   
        model.fit(x_train,y_train)
        score=model.score(x_train,y_train)
        score1=model.score(feature_testDF,target_test)
        pred=model.predict(x_train)
        pred1=model.predict(feature_testDF)
        class1=classification_report(y_train,pred)
        class2=classification_report(target_test,pred1)
        print(f'모델: {x}')
        print(f'train score: {score}, testscore: {score1}')
        print('train report')
        print(class1)

        print('test report')
        print(class2)
        print('*'*40)
    
    except Exception:
        pass
```

```python
model=HistGradientBoostingClassifier(early_stopping=True)
param_grid={'learning_rate':[0.01,0.001,0.1],'max_iter':[100,200,300],'max_leaf_nodes':[31,41,51],'max_depth':range(1,10)}
grid=GridSearchCV(estimator=model,param_grid=param_grid)
grid.fit(x_train,y_train)
print(grid.best_estimator_)
print(grid.best_score_)
# score=model.score(x_train,y_train)
# score1=model.score(x_test,y_test)
# pred=model.predict(x_train)
# pred1=model.predict(x_test)
# class1=classification_report(y_train,pred)
# class2=classification_report(y_test,pred1)
# print(f'train score: {score}, testscore: {score1}')
# print('train report')
# print(class1)
# print('test report')
# print(class2)
```

```python
model=KNeighborsClassifier(n_neighbors=5,weights='distance',algorithm='kd_tree',p=2)
model.fit(x_train,y_train)
score=model.score(x_train,y_train)
score1=model.score(x_test,y_test)
pred=model.predict(x_train)
pred1=model.predict(x_test)
class1=classification_report(y_train,pred)
class2=classification_report(y_test,pred1)
print(f'train score: {score}, testscore: {score1}')
print('train report')
print(class1)
print('test report')
print(class2)
```
