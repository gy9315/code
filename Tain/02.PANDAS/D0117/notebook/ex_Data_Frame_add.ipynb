<hr> [3] 행 추가 <hr>

```python
## 변수명.loc(추가할 행 인덱스)=값
memberDF.loc[7]=100
memberDF
```

```python
## 변수명.loc[추가 행 인덱스]=[col 수]
## 컬럼수 와 데이터 수가 다르면 ERROR
memberDF.loc[8]=[9,'태권브이',100,'대구']
memberDF

# memberDF.loc[9]=[8,'태권브이']
```

```python
memberDF.loc[8]=0
memberDF
memberDF['비고']=0
memberDF
memberDF.loc[7,'비고']=100

```

[4] 열 추가 <hr>

```python
# 변수명[컬럼명]=값
memberDF['성별']='남'
memberDF
# memberDF['성별'].loc[1::2]='여'
## 규칙적인 숫자가 있으면 전체 선택저긍로 변경가능능
memberDF.loc[memberDF.index[::2],'번호':'이름':2]='여'
memberDF
```

```python
memberDF.loc[:,'비고']=5
memberDF
```

```python
memberDF.loc[(memberDF['나이']>=20) & (memberDF['지역']=='대구'),'비고']=1
memberDF
```
