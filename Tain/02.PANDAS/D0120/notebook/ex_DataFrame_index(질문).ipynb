[1] 모듈 로딩 및 데이터 준비 <hr>

```python
# [1-1] module 로딩
import pandas as pd
```

```python
# [1-2] 데이터 준비
path='../DATA/member.csv'
```

[2] 데이터 로딩 및 확인


```python
mDF=pd.read_csv(path,header=[0])
print(mDF)
mDF.set_index(['번호','이름'])
```

[3] 인덱스 다루기

```python
# [3-1] 인덱스와 col을 서로 바꾸기(Transpose)
# 2가지 방법
# - method: transpose()
# - 속성: T
mDF.transpose()
mDF.T
mDF
```

```python
# [3-2] 컬럼을 행 인덱스 설정하기
# - 조건: 컬럼데이터로 한 행의 데이터 추출 가능하는 컬럼(중복이 없는)
# - method: set_index(컬럼명) or (['col1','col2',...])
# - 이름 컬럼을 행 인덱스로 설정
mDF2=mDF.set_index(['이름','지역'])
print(mDF2.columns)
print(mDF2)
mDF2.reset_index()

```

```python
# 행 선택
mDF2.loc[['이나영','박지민']]
```

```python
# 여러행 선택 # multi index slicing 방법:loc()
mDF2
mDF2.iloc[0:4]
mDF2.index
print(mDF2)
print(mDF2.loc[['홍길동','마징가']])
print(mDF2.index)
# print(mDF2.loc['홍길동':'마징가'])
print(mDF2.iloc[0:4])
# print(mDF2.loc[('홍길동','대구'):('마징가','부산')])
mDF2=mDF2.reset_index(level=1)
print(mDF2)
mDF2.index
mDF2.loc['홍길동':'최희진']
mDF2=mDF2.reset_index()
mDF2.iloc[0:2]
mDF2=mDF2.set_index(['이름','지역'])
mDF2.iloc[:3]
```

```python
mDF2.sort_index()
# mDF2.sort_values(by=['이름','지역'],axis=1) # multi index기준으로 sort 하는 법
mDF2
```
