```python
# [1-1] 모듈로딩
import pandas as pd # 데이터 분석 및 전처리 모듈
import matplotlib.pyplot as plt # 시각화 모듈
```

```python
# [1-2] 데이터 준비
data={'A':[10,20,30,40,50],'B':[15,25,35,45,55]}
dataDF=pd.DataFrame(data)
```

[데이터 시각화] - 선 그래프 <hr>

```python
# plt.plot(ydata) 또는 plt.plot(xdata, ydata)
# A컬럼 시각화

plt.plot(dataDF.index,dataDF['A'])
plt.plot(dataDF)
```

```python
# 선X, mark로 표현
# plt.plot(dataDF['A'],'o') # => 기본색
# plt.plot(dataDF['A'],'ro') # => 빨간색
# plt.plot(dataDF['A'],'^ ') # => 빨간색
plt.plot(dataDF['A'],'m^') # => 자주색
plt.show()
```

```python
# 선X, mark로 표현
# plt.plot(dataDF['A'],'o') # => 기본색
# plt.plot(dataDF['A'],'ro') # => 빨간색
# plt.plot(dataDF['A'],'^ ') # => 빨간색
# plt.plot(dataDF['A'],'m*--') # => 자주색 이면서 --스타일 선
plt.plot(dataDF['A'],'mo:') # => 자주색 이면서 o마커 --스타일 선
plt.show()
```

```python
# 선X, mark로 표현
# plt.plot(dataDF['A'],'mo:') # => 자주색 이면서 --스타일 선
plt.plot(dataDF['A'],'mo--',lw=3.0) # => 자주색 이면서 --스타일 선, 라인두께
plt.plot(dataDF['A'],'o--m',lw=3.0,ms=50.0) # => 자주색 이면서 --스타일 선
plt.show()
```

```python
## A컬럼과 B컬럼 시각화
xdata=dataDF.index+1
ydata1=data['A']
ydata2=data['B']
plt.plot(xdata,ydata1)
plt.plot(xdata,ydata2)
```

```python
## A컬럼과 B컬럼 시각화 + 범례 <- 여러개 그래프 식별별
xdata=dataDF.index+1
ydata1=data['A']
ydata2=data['B']
plt.plot(xdata,ydata1,label='A col') # A컬럼
plt.plot(xdata,ydata2,label='B col') # B컬럼
plt.legend(loc='lower right') # 그래프별 라벨 설정
plt.title("A col compare to B col")
plt.xlabel("index",fontdict={'color':'r'})
plt.ylabel("value")
plt.grid()
plt.show()
```
