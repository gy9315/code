[Matplotlib의 히스토그램]
- 데이터의 분포 체크를 위한 시각화 그래프
- 이상치/패턴 파악 가능

```python
# [1-1] 모듈 로딩
import matplotlib.pyplot as plt
import numpy as np
import random as rd
import pandas as pd
```

```python
np.random.seed(3)
data=np.random.randn(1000)
print(f'평균{data.mean():.1f}, 표준편차: {data.std():.1f}')
plt.figure(figsize=(8,8)).set_facecolor('lightgray')
plt.hist(data,bins=100,color='y',edgecolor='k')
plt.legend()
```

```python
np.random.seed(64)
data=np.random.randn(1000)
print(f'평균{data.mean():.1f}, 표준편차: {data.std():.1f}')
plt.figure(figsize=(8,8)).set_facecolor('lightgray')
plt.hist(data,bins=100,color='y',edgecolor='k',histtype='step')
plt.legend()
```

```python
data2 = (np.random.randn(1000) + 2) * 1.5  
print(f'평균:{data2.mean():.1f}, 표준편차:{data2.std():.1f}')
## 시각화
## density 매개변수 : y축을 비율로 표현  
plt.hist(data, label='Data1', density=True, alpha=0.75, edgecolor='k')  
(plt.hist(data2,label='Data2', density=True, alpha=0.75, edgecolor='k')  
```
