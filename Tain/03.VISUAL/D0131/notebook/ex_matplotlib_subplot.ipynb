```python
DF=pd.read_csv('../../02.PANDAS/DATA/auto_mpg.csv')
```

```python
# [여러개 그래프 그리기]
fig=plt.figure(figsize=(10,10))
fig.set_facecolor('yellow')
plt.suptitle('mpg',size=20)
plot1=plt.subplot(2,2,1)
plot2=plt.subplot(222)
plot3=plt.subplot(223)
plot4=plt.subplot(224)
plot1.plot(range(len(DF)),DF['mpg'],'o')
plot1.set_xticks(range(0,len(DF.index),50),DF.index[::50])
plot1.set_title('mpg')
plot2.bar(DF.index,DF['mpg'])
plot2.set_title('mpg')
plot3.boxplot(DF['mpg'])
plt.tight_layout()

plt.show()
```

```python
# 여러개 그래프 그리기
## - 방법 2
## - subplots()
## 0번 행이 있음
figs,axes=plt.subplots(1,2)
print(axes)
axes[0].plot(DF['mpg'])
```
