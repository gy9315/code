```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# 내용 정하기
ratio=[10,20,30]
# 각 내용의 label정하기
label=['apple','banana','kiwi']
# 색깔 정하기
colors=['royalblue','tomato','green']
# 중심에서 0.1씩 뛰우기
explodes=[0.1,0.1,0.1]
# pie 그리기
# 도넛 모양 만들기
wedgeprops={'width':0.7,'edgecolor':'w','linewidth':1}
plt.pie(ratio,labels=label,colors=colors,explode=explodes,autopct='%1.1f%%',shadow=True,wedgeprops=wedgeprops)
plt.legend(loc=1,bbox_to_anchor=(1.3,1))
```

```python
# 내용 정하기
ratio=[10,20,30]
# 각 내용의 label정하기
label=['apple','banana','kiwi']
# 색깔 정하기
colors=['royalblue','tomato','green']
# 중심에서 0.1씩 뛰우기
explodes=[0.1,0.1,0.1]
# pie 그리기
# 도넛 모양 만들기
wedgeprops={'width':0.7,'edgecolor':'w','linewidth':1}
# 값 출력 위치 바꾸기
plt.pie(ratio,labels=label,colors=colors,explode=explodes,autopct='%1.1f%%',shadow=True,wedgeprops=wedgeprops,pctdistance=1.2,labeldistance=0.5)
plt.legend(loc=1,bbox_to_anchor=(1.3,1))
```
