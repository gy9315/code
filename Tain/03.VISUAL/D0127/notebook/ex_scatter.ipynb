```python
# [2-2] auto mpg 파일 기반 산점도
# mpg와 weight 컬럼의 관계성
pd1=pd.read_csv('../../02.PANDAS/DATA/auto_mpg.csv')
# pd1.info()
mpgDF=pd1.iloc[:,0]
wDF=pd1.iloc[:,4]
plt.scatter(mpgDF,wDF,label='MPG와 Weight 상관관계',s=50,c='y')
plt.legend()
plt.ylabel('kg',rotation='horizontal')
plt.xlabel('Mile/Gal')
# wDF.unique()
# mpgDF.unique()
s1=round(mpgDF.corr(wDF),2)
plt.text(mpgDF.values.mean(),wDF.values.mean(),s=f'상관계수: {s1}',size=30,ha='left',va='center' )
```

```python
# [2-2] auto mpg 파일 기반 산점도
# mpg와 weight, cylinder 컬럼의 관계성
pd1=pd.read_csv('../../02.PANDAS/DATA/auto_mpg.csv')
pd1.info()
mpgDF=pd1.iloc[:,0]
wDF=pd1.iloc[:,4]
cDF=pd1.iloc[:,1]
plt.scatter(mpgDF,wDF,label='MPG/Weight/cylinder 상관관계',s=cDF*5,c='y')
plt.legend()
plt.ylabel('kg',rotation='horizontal')
plt.xlabel('Mile/Gal')
# wDF.unique()
# mpgDF.unique()
s1=round(mpgDF.corr(wDF),2)
plt.text(mpgDF.values.mean(),wDF.values.mean(),s=f'상관계수: {s1}',size=30,ha='left',va='center' )
```

```python
pd1['origin'].unique()
oDF=pd1['origin']
plt.scatter(mpgDF,wDF,label=['MPG/Weight/cylinder 상관관계',2,3],s=cDF*5,c=oDF)
plt.colorbar(ticks=[3,2,1])

plt.legend()
plt.ylabel('kg',rotation='horizontal')
plt.xlabel('Mile/Gal')
# wDF.unique()
# mpgDF.unique()
s1=round(mpgDF.corr(wDF),2)
plt.text(mpgDF.values.mean(),wDF.values.mean(),s=f'상관계수: {s1}',size=30,ha='left',va='center' )
```
