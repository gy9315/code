[matplotlib - 영역 그래프]
- 선 그래프에서 특정부분에 색상 채우기
- 너무 조각 조각 내면 오히려 가독성 떨어짐
- 강조하고 싶은 부분에 설정
- plot() 선 그래프 그리기 -> fill_between()/fill() 영역 색상지정 <hr>

```python
xdata=pd.Series([1,2,3,4])
ydata=pd.Series([2,3,5,10])
plt.plot(xdata,ydata)
plt.xlabel('x축')
plt.ylabel('y축',rotation='horizontal')
ydata1=pd.Series([1,2,4,8])
plt.plot(xdata,ydata1)
plt.fill_between(xdata[2:4],ydata[2:4],ydata1[2:4],alpha=0.2)
plt.text(3.5,7,s='접점')
```

```python
xdata=pd.Series([1,2,3,4])
ydata=pd.Series([2,3,5,10])
plt.plot(xdata,ydata)
plt.xlabel('x축')
plt.ylabel('y축',rotation='horizontal')
ydata1=pd.Series([1,2,4,8])
plt.plot(xdata,ydata1)
# 왼쪽 아래 꼭지점 부터 좌표 채우기기
plt.fill([2,2,3,3,2.5],[2,4,6,3,2],alpha=0.2)
```
