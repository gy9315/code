```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import font_manager as fm
from matplotlib import rc
font_name=fm.FontProperties(fname='C:\Windows\Fonts\HancomHoonminjeongeumH.ttf').get_name()
rc('font', family=font_name)
```

```python
path='../DATA/남북한발전전력량.xlsx' 
DF1=pd.read_excel(path)
DF2=pd.read_excel(path)
```

```python

DF1.replace('-',0,inplace=True)
# # pd.concat([DF1, ]
y=['북한','신재생']
for x in range(27):
    y.append(0)
DF1.loc[9]=y
# len(DF1.columns)
DF1.ffill(inplace=True)
DF1
DF1.iloc[:,2:]=DF1.iloc[:,2:].astype('int64')
# [4-1] 수력 값만 출력력
DF1.set_index(list(DF1.columns[:2]),inplace=True)
data=DF1.xs('수력',level=1).iloc[0].values
data1=DF1.xs('수력',level=1).iloc[1].values
data
list1=list(range(0,len(DF1.columns),6))
list2=[DF1.columns[i] for i in list1]
plt.figure(figsize=(5,5))
plt.xticks(list1,labels=list2)
plt.bar(DF1.columns,data1,alpha=0.2)
plt.bar(DF1.columns,data)

```

```python
# xdata=2개
# ydata=2개
xdata=pd.Series(range(len(DF1.columns)))
plt.bar(xdata,data,width=0.5,label='북한',alpha=0.5)
plt.bar(xdata+0.5,data1,width=0.5,label='남한')
# xticks의 숫자값은 x축의 절대적인 숫자 값
plt.xticks(xdata[::5]+0.25,labels=DF1.columns[::5])
plt.legend()
plt.title('수력발전량',pad=15)
plt.xlabel('연도')
plt.ylabel('발전량',rotation='horizontal')
```
